```python
import folium

from folium.plugins import minimap
Chernobyl = folium.Map(location = (51.38972265969885, 30.093096163800222))
Chernobyl

```

```python
from folium.plugins import MiniMap
from branca.element import Figure

popuptext = '<b>Чорнобильська атомна електростанція<b>'

my_map = folium.Map(location=[51.38972265969885, 30.093096163800222],zoom_start=16,tiles='stamenterrain')

folium.Marker(location=[51.38972265969885, 30.093096163800222],popup=popuptext,icon=folium.Icon(color="green",icon="ok-sign")).add_to(my_map)

folium.Circle(location=[51.38972265969885, 30.093096163800222],color='black',weight=6, fill_color="yellow",fill_opacity=0.5,tooltip="Чорнобильська атомна електростанція").add_to(my_map)

minimap=MiniMap(tile_layer='stamenterrain',position='bottomright')

my_map.add_child(minimap)

my_map.save('C:folium.html')

my_map

```

```python
fig = Figure(width=1000, height=500)
fig.add_child(my_map)
fig
```
