```python
import pandas as pd
import matplotlib.pyplot as plt

datos = pd.read_csv("casasboston.csv")

print(datos)

df = datos[["RM", "CRIM", "MEDV", "TOWN", "CHAS"]]

df = datos.rename(columns={
    "TOWN":"CIUDAD",
    "CRIM":"INDICE DE CRIMEN",
    "INDUS":"PCT_ZONA_INDUSTRIAL",
    "CHAS":"RIO CHARLES",
    "RM":"N_HABITACIONES_MEDIO",
    "MEDV": "VALOR_MEDIANO",
    "LSTAT":"PCT_CLASE_BAJA"
})

print(df)


```

```python
df.N_HABITACIONES_MEDIO.plot.hist()
plt.show()
```

```python

df.plot.scatter( x="INDICE DE CRIMEN" ,y="VALOR_MEDIANO", alpha=0.5)
plt.show()
```

```python
valor_por_ciudad=df.groupby("CIUDAD")["VALOR_MEDIANO"].mean()

valor_por_ciudad.head(10).plot.barh()
plt.show()
```

```python
import seaborn as sns
table =pd.crosstab(df['INDICE DE CRIMEN'], df['VALOR_MEDIANO'])
print(table)
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(table,annot=True,cmap="coolwarm")
plt.title('Mapa Indice y valor medio')
plt.show()
```
