```python
from random import random
import matplotlib.pyplot as plt
import numpy as np
import math as mt

constante_gravitacional = 6.67430e-11
#Nota: r = distancia

class Fuerza_Gravitacional():
  def __init__(self, masa1, masa2, distancia):
    self.masa1 = masa1
    self.masa2 = masa2
    self.distancia = distancia

  def calcular_fuerza_gravitacional(self, masa1, masa2, distancia):
      fuerza = constante_gravitacional * (self.masa1 * self.masa2) / mt.pow(self.distancia, 2)
      return fuerza

  def calcular_aceleracion(self, masa_cuerpo):
    return constante_gravitacional*masa_cuerpo / (mt.pow(self.distancia, 2))

    def ingresar_datos():
      masa1 = float(input("Ingrese la masa del objeto 1 (kg): "))
      masa2 = float(input("Ingrese la masa del objeto 2 (kg): "))
      distancia = float(input("Ingrese la distancia entre los objetos (m): "))
      fuerza=calcular_fuerza_gravitacional(masa1, masa2, distancia)
      print("La fuerza gravitacional es:", {"fuerza:.2e"}, "N")


    def main():
        ingresar_datos()

    if __name__ == "__main__":
        main()


      #ingresar_datos()
```
